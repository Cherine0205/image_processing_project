```python
import nibabel as nib
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import os, shutil
```

```python
image_input='C:/Users/24710/Downloads/B/images'
label_input='C:/Users/24710/Downloads/B/labels'
list_f=[os.path.join(image_input,f) for f in os.listdir(image_input)]
slices=[nib.load(f) for f in list_f] #slices中为1-6号的图像（320,320,x）
list_l=[os.path.join(label_input,f) for f in os.listdir(label_input)]
labels=[nib.load(l) for l in list_l]#labels中为1-6号的label（320,320,x）
```

```python
input_arr=[slices[i].get_fdata() for i in range(6)]
[input_arr[0].shape,input_arr[1].shape,input_arr[2].shape,input_arr[3].shape,input_arr[4].shape,input_arr[5].shape]
```

```python
train_data=np.ones([450,320,320,1])
```

```python
n=0
for i in range(4):
    x,y,z=input_arr[i].shape
    for j in range(z):
        a=input_arr[i][:,:,j]
        b=np.atleast_3d(a)
        #b=a
        #c=b[np.newaxis,:]
        b = b.reshape((1,) + b.shape) 
        train_data[n,:,:,:]=b
        #train_data[n,:,:]=c
        n=n+1
```

```python
input_label=[labels[i].get_fdata() for i in range(6)]
train_label=np.ones([450,320,320,1])
```

```python
n=0
for i in range(4):
    x,y,z=input_label[i].shape
    for j in range(z):
        a=input_label[i][:,:,j]
        b=np.atleast_3d(a)
        #b=a
        #c=b[np.newaxis,:]
        b = b.reshape((1,) + b.shape) 
        train_label[n,:,:,:]=b
        #train_label[n,:,:]=c
        n=n+1

train_label=train_label*65535
```

```python
train_data=train_data[:,100:250,100:250,:]
train_label=train_label[:,100:250,:100:250,:]
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
'''datagen=ImageDataGenerator(rescale=1./65535,rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)'''
```

```python
datagen=ImageDataGenerator(rescale=1./65535)
```

```python
train_generator=datagen.flow(train_data,train_label, batch_size=1)
```

```python
history=model.fit_generator(train_generator,epochs=10,steps_per_epoch=45)
```

```python
img=input_arr[4][:,:,60]
img=img.reshape((1,)+img.shape)
result=model.predict(img)
```
