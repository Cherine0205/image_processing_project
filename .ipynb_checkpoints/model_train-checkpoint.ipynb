```python
for i in range(5):
    if i ==2:
        n=40
        for j in range(40,60,1):
            a=input_arr[i][:,:,j]
            #a=a*1.0/2095
            b=np.atleast_3d(a)
            #b=a
            #c=b[np.newaxis,:]
            b = b.reshape((1,) + b.shape) 
            train_data[n,:,:,:]=b
            
            x=input_label[i][:,:,j]
            y=np.atleast_3d(x)
            y=y.reshape((1,)+y.shape)
            train_label[n,:,:,:]=y
            n=n+1
    if i ==3:
        n=80
        for j in range(60,80,1):
            a=input_arr[i][:,:,j]
            #a=a*1.0/2095
            b=np.atleast_3d(a)
            #b=a
            #c=b[np.newaxis,:]
            b = b.reshape((1,) + b.shape) 
            train_data[n,:,:,:]=b
            
            x=input_label[i][:,:,j]
            y=np.atleast_3d(x)
            y=y.reshape((1,)+y.shape)
            train_label[n,:,:,:]=y
            n=n+1
    if i ==4:
        n=100
        for j in range(60,80,1):
            a=input_arr[i][:,:,j]
            #a=a*1.0/2095
            b=np.atleast_3d(a)
            #b=a
            #c=b[np.newaxis,:]
            b = b.reshape((1,) + b.shape) 
            train_data[n,:,:,:]=b
            
            x=input_label[i][:,:,j]
            y=np.atleast_3d(x)
            y=y.reshape((1,)+y.shape)
            train_label[n,:,:,:]=y
            n=n+1
    if i == 0:
        n=0
        for j in range(60,80,1):
            a=input_arr[i][:,:,j]
            #a=a*1.0/2095
            b=np.atleast_3d(a)
            #b=a
            #c=b[np.newaxis,:]
            b = b.reshape((1,) + b.shape) 
            train_data[n,:,:,:]=b
            
            x=input_label[i][:,:,j]
            y=np.atleast_3d(x)
            y=y.reshape((1,)+y.shape)
            train_label[n,:,:,:]=y
            n=n+1
    if i==1:
        n=20
        for j in range(60,80,1):
            a=input_arr[i][:,:,j]
            #a=a*1.0/2095
            b=np.atleast_3d(a)
            #b=a
            #c=b[np.newaxis,:]
            b = b.reshape((1,) + b.shape) 
            train_data[n,:,:,:]=b
            
            x=input_label[i][:,:,j]
            y=np.atleast_3d(x)
            y=y.reshape((1,)+y.shape)
            train_label[n,:,:,:]=y
            n=n+1
            
            
train_label=train_label
train_data=train_data
[train_data.shape,train_label.shape]
```
